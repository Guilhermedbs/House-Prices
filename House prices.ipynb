## Analyzing the dependent variable

```python
import scipy.stats as stats

mu, sigma = stats.norm.fit(df['SalePrice'])
```

```python
sns.set_theme(style='darkgrid', palette='icefire')
```

```python
# Ploting the histogram of the Sales Price to analyze it's distribuition.
plt.figure(figsize=(12,6))
sns.histplot(df['SalePrice'], kde=True, bins = 50, stat='density')
plt.title('Sales Prices Real Distribuition', fontsize=20)
plt.xlabel('Sales Prices', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

```python
# Creating the values of a normal distribuition for the same values of mean(mu) and standard deviation(sigma)
x_norm = np.linspace(df['SalePrice'].min(), df['SalePrice'].max(), 100)
y_norm = stats.norm.pdf(x_norm, loc=mu, scale=sigma)
```

```python
# Ploting a scatter plot of how a normal distribution would be for those values
plt.figure(figsize=(12,6))
sns.scatterplot(x=x_norm, y=y_norm)
plt.title('Sales Prices Normal Distribuition', fontsize=20)
plt.xlabel('Sales Prices', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

```python
# Comparing the too graphs to see if the dependent variable fits a normal distribuition
# or if it needs to be normalized

plt.figure(figsize=(12,6))
sns.histplot(df['SalePrice'], bins = 50, stat='density', label='Real Distribuition')
sns.scatterplot(x=x_norm, y=y_norm, label='Normal Distribuition')
plt.title('Sales Prices Real X Normal Distribuition', fontsize=20)
plt.xlabel('Sales Prices', fontsize=16)
plt.ylabel('Density', fontsize=16)
plt.legend(fontsize=14)
```

```python
qq_data = stats.probplot(df['SalePrice'], dist='norm')
slope, intercept, r_value, p_value, std_err = stats.linregress(qq_data[0][0], qq_data[0][1])
line_x = np.array(qq_data[0][0])
line_y = intercept + slope * line_x
```

```python
# Using a Q-Q plot to compare the real data quantiles with the quantiles that would 
# be expected in a normal distribution. 
plt.figure(figsize=(12,6))
sns.scatterplot(x=qq_data[0][0], y=qq_data[0][1], color='b', label='Real') 
sns.lineplot(x=line_x, y = line_y, label='Normal')
plt.title('Q-Q plot', fontsize=20)
plt.xlabel('Theorical Quantiles', fontsize=16)
plt.ylabel('Sales Prices', fontsize=16)
plt.legend(fontsize=14)
```

## 1.Distribution of dwelling types and their relation to sale prices

```python
# Distribution of building types and how it affects the Sale Price
plt.figure(figsize=(12,6))
building_types = df['BldgType'].value_counts()
ax = sns.barplot(x=building_types.index, y=building_types.values )
ax.set_title('Distribution of Building Types', fontsize = 20)
ax.set_xlabel('Building Types', fontsize=16)
ax.set_ylabel('Count', fontsize=16)
ax.bar_label(ax.containers[0], fontsize=10);
```

```python
building_prices = df.groupby('BldgType')['SalePrice'].mean()
plt.figure(figsize=(12,6))
ax = sns.barplot(x=building_prices.index, y=building_prices.values)
ax.set_title('Avarege Sale Price by Building Type', fontsize = 20)
ax.set_xlabel('Building Types', fontsize=16)
ax.set_ylabel('Sale Price', fontsize=16)
for position, value in enumerate(building_prices.values):
    ax.text(position, value, f"${value:.0f}", ha='center', va='bottom', fontsize=12)
```

## 5.Is there a Correlation between Property Age and Sale Price

```python
# Calculating property age
df['property_age'] = df['YrSold'] - df['YearBuilt']

#Calculating the Correlation between Property age and Sale Price
age_price_corr = df['property_age'].corr(df['SalePrice'])
print(f'Correlation Between Property Age and Sale Price: { age_price_corr}')

# Ploting a Scatter with a regression line plot to visualize how age affecets Sale Price.
plt.figure(figsize=(12,6))
sns.pairplot(data=df, x_vars=['property_age'], y_vars=['SalePrice'],height = 6, aspect=2, kind='reg')
plt.title('Property Age x Sale Price', fontsize=20)
plt.xlabel('Property Age',fontsize=16)
plt.ylabel('Sale Price',fontsize=16)
```

## 6.Is there a Correlation between Living Area and Sale Price


```python
#Calculating the Correlation between Property Living Area and Sale Price
living_area_price_corr = df['GrLivArea'].corr(df['SalePrice'])
print(f'Correlation Between Property Living Area (Above Grade) and Sale Price: {living_area_price_corr}')

# Ploting a Scatter with a regression line plot to visualize how Living Area affecets Sale Price.
plt.figure(figsize=(12,6))
sns.pairplot(data=df, x_vars=['GrLivArea'], y_vars=['SalePrice'],height = 6, aspect=2, kind='reg')
plt.title('Property Living Area (above grade) x Sale Price', fontsize=20)
plt.xlabel('Living Area',fontsize=16)
plt.ylabel('Sale Price',fontsize=16)
```

## 7.Does price change year to year?

```python
# Box plot of price over the years
average_year_prices = df.groupby('YrSold')['SalePrice'].mean()
plt.figure(figsize=(12,6))
ax =sns.boxplot(data=df, x='YrSold', y='SalePrice')
ax.set_title('Yearly Distribution of Sale Prices',fontsize=20)
plt.xlabel('Year',fontsize=16)
plt.ylabel('Sale Price',fontsize=16)
for position, value in enumerate(average_year_prices.values):
    ax.text(position, value, f"${value:.0f}", ha='center', va='bottom', fontsize=12, color='white')

```
