```python
# Creating a function that generates a bar plot to visually represent the distribution 
# of missing data in a DataFrame, providing insights into which columns have the most missing values. 
def missing_data(df):
    missing_data_size = df.isnull().sum()
    missing_data_columns = (
                  (missing_data_size[missing_data_size > 0]
                  ).sort_values(ascending=False).to_frame(name='Count')
      )
    missing_data_columns = (
                  missing_data_columns.assign(
                  Column=missing_data_columns.index
                  )
      ).reset_index(drop=True)

    plt.figure(figsize=(12, 6))
    ax = sns.barplot(data=missing_data_columns, x='Count', y='Column',palette='icefire')
    ax.bar_label(ax.containers[0], labels=[f"{i:.2f}%" for i in missing_data_columns.Count.astype(int) / len(df) * 100])
    plt.title("Missing data distribuition", fontsize=20)
    plt.xlabel('Total amount', fontsize=16)
    plt.ylabel('Columns names', fontsize=16)
    plt.show()
```

## Data Preprocessing

```python
# Creating a normalized dependent variable
log_saleprice = np.log(df['SalePrice'])
# Creating histplot to see if a normalized variable fits better a normal distribuition
plt.figure(figsize=(12,6))
sns.histplot(log_saleprice, bins = 50, stat='density', kde=True , label='Real Distribuition')
plt.title('Sale Price(Log - Transformed) Distribuition', fontsize=20)
plt.xlabel('Sale Price(Log-Transformed)',fontsize=16)
plt.ylabel('Density',fontsize=16)
```

```python
log_saleprice
```

```python
# MSSubClass is a categorical feature that Identifies the type of dwelling involved in the sale.	
df['MSSubClass'].dtype
# But here it is indentifies as numerical feature
```

```python
#Changing MSSubClass to categorical feature
df['MSSubClass'] = df['MSSubClass'].astype('category')
df['MSSubClass'].dtype
```

```python
# Crating a Data Pipeline

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# Defining Transformers for categorical and numerical columns using Pipelines
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy ='mean')),
    ('scaler', MinMaxScaler())
    ])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
     ('onehot', OneHotEncoder(handle_unknown='ignore', sparse=False))
    ])

# Defining categorical and numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Droping the dependent variable
numerical_columns = numerical_columns.drop('SalePrice')
```

```python
from sklearn.compose import ColumnTransformer

# Using Column Transforme to aply both categorical and numerical transformations at the same time
preprocessor = ColumnTransformer(transformers =[
                ('numerical', numerical_transformer, numerical_columns),
                ('categorical', categorical_transformer, categorical_columns),
                ],remainder= 'passthrough')

# Creating pipeline with the preprocessor
pipe = Pipeline(steps=[('preprocessor', preprocessor)])
```

```python
# Aplying the pipeline to the dataset
X = df.drop('SalePrice', axis=1)
y = log_saleprice
X_preprocessed = pipe.fit_transform(X)
```

```python
X_preprocessed.shape
```

```python
# Crating a dataset with the transformed data 
columns = list(numerical_columns) + list(preprocessor.named_transformers_['categorical']['onehot'].get_feature_names_out(categorical_columns))
df_trans = pd.DataFrame(X_preprocessed, columns=columns, index=df.index)
df_trans
```

```python
X = df_trans
```

## Creating models

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn import metrics

# Spliting the dataset to train the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Defining the model that are going to be evaluated
models = {
    'Linear_Regression': LinearRegression(),
    'Random_Forest': RandomForestRegressor(random_state=42),
    'Gradient_Boosting': GradientBoostingRegressor(random_state=42)
}

# Defining hyperparameters grid

param_grid = {
    'Linear_Regression': {},
    'Random_Forest': {'n_estimators': [50, 100, 150], 'max_depth': [None, 10, 20]},
    'Gradient_Boosting': {'n_estimators': [50, 100, 150], 'learning_rate': [0.01, 0.1, 0.2]}
}
```

```python
# Creating a dictionary to stoe the trained models
trained_models = {}

# Evaluating models using GridSearhCV and printing the results

for model_name, model in models.items():
    grid_search = GridSearchCV(model, param_grid[model_name], cv=5, scoring='r2')
    grid_search.fit(X_train, y_train)
    
    print(f'Model: {model_name}')
    print(f'Best Parameters: {grid_search.best_params_}')
    
    trained_models[model_name] = grid_search.best_estimator_
    
    # Evaluating the model with the test set
    score = grid_search.score(X_train,y_train).round(3)
    y_pred = grid_search.predict(X_test)
    r2 = metrics.r2_score(y_test, y_pred)
        
    print(f'Estimated R-Squared score on Training Set {score.round(3)}')
    print(f'R-Squared score on Test Set {r2.round(3)}')
    print('-'*50)
    
```

## Analyzing the Models 

### Linear Regression Model

```python
# The R-squared scores for the Linear Regression model suggest a potential issue with overfitting
# Because of that a more indepth analysis will be needed
from sklearn.model_selection import learning_curve

# Creating a function to visualize the learning curve of the model
def plot_learning_curve(model, X, y):
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5, scoring='r2', train_sizes=np.linspace(0.1, 1.0, 10))
    
    # Calculating the score means for the cross-validation
    train_scores_mean = np.mean(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    
    plt.figure(figsize=(12, 6))
    plt.plot(train_sizes, train_scores_mean, 'o-' , label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-',label="Cross-validation score")
    plt.title(f"Learning Curve for {model.__class__.__name__}",fontsize=20)
    plt.xlabel('Training examples',fontsize=15)
    plt.ylabel('R-squared score',fontsize=15)
    plt.legend(fontsize=13)
    sns.despine()
    plt.show() 
```

```python
# Visualizing the model's learning curve
plot_learning_curve(trained_models['Linear_Regression'], X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score

# Checking the cross-validation scores with cross_val_score
scores = cross_val_score(trained_models['Linear_Regression'], X_train, y_train, cv=5, scoring='r2')
print("Cross-validated R-squared scores:", scores)
print("Cross-validated R-squared mean score:", (scores.mean()).round(3))
```

```python
# Crating a function too ajust the new data set
def ajusting_transformed_dataset(X_original, X_new, rows_num):
    """ 
    Adjusts the dataframe for model input after the pipeline transformation.

    Parameters:
    -----------
    X_original (pd.DataFrame): The original dataset used for training the model after the pipeline transformation.
    X_new (pd.DataFrame): The new dataset that will be inputted into the model after the pipeline transformation.
    rows_num (int): The number of rows in the new dataset.

    Returns:
    --------
    pd.DataFrame: Adjusted dataframe for model input."""
    missing_columns = list(set(X_new.columns).difference(X_original.columns))
    for column in missing_columns:
        X_new.drop(column,inplace=True,axis=1)
    
    missing_columns = list(set(X_original.columns).difference(X_new.columns))
    for column in missing_columns:
        X_new[column] = np.zeros(rows_num)
    return X_new
```
